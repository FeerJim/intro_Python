```python
import matplotlib.pyplot as plt


labels = ['G1', 'G2', 'G3', 'G4', 'G5'] # Etiquetas
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35       # Ancho de las barras

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, yerr=men_std, label='Hombres')
ax.bar(labels, women_means, width, yerr=women_std, bottom=men_means,
       label='Mujeres')

ax.set_ylabel('Puntajes')
ax.set_title('Puntajes por género')
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(labels, men_means, width, yerr=men_std, label='Hombres', color = 'c')
ax.bar(labels, women_means, width, yerr=women_std, bottom=men_means,
       label='Mujeres',color = 'm')

ax.set_ylabel('Puntajes')
ax.set_title('Puntajes por género')
ax.legend()

plt.show()
```

### Excepciones

Nos permiten evitar que el codigo se interrumpa, para ello usaremos las sentencias `try`, `except` y `finally`.

```python
# sentencia try-except
try:
    1/0
except:
    print('Error: No se puede dividir para cero.')
```

Que sucede si queremos escribir diferentes tipos de errores..??

```python
try:
    # escribimos nuestro codigo aqui
except ZeroDivisionError:
    # escribimos el mensaje que se debe visualizar
except:
    # Escribimos un mensaje en caso de que sea una error que no hemos expecificado antes
```

```python
try:
    1/0
except(RuntimeError, TypeError, NameError):
    pass
except:
    print('Error: No se puede dividir para cero.')
```

```python
try:
    1/0
except ZeroDivisionError as detalle:
    print('Error: Esto no es posible:',detalle)
```

**Ejemplo:**
Error de tipo al pedir un valor de pantalla.

```python
try:
    x = float(input('Ingrese un numero: '))
except:
    print('Error: No ha ingresado un numero valido.')
```

```python
while True:
    try:
        x = float(input('Ingrese un numero positivo: '))
    except:
        print('Error: No ha ingresado un numero valido. Intente otra vez :-)')
    if x > 0:
        break
    print('Error: El número ingresado no es positivo :c. Intenta otra vez :-)')
```

Cuando usamos `except` no es necesario usar `finally` al final, sin embargo podemos usarla.

```python
# sentencia try-except-finally
try:
    x = float(input('Ingrese un numero: '))
except:
    print('Error: No ha ingresado un numero valido.')
finally:
    print('Gracias :3')
```

Usando la sentencia `raise` para crear excepciones

```python
try:
    raise Exception('Error 1', 'Error 2')
except Exception as inst:
    print(type(inst))     # la instancia de excepción
    print(inst.args)      # argumentos guardados en .args
    print(inst)           # __str__ permite imprimir args directamente
    x = inst          # __getitem__ permite usar args directamente
    print('x =', x)
```

```python
def dividir(dividendo, divisor):
    try:
        resultado = dividendo / divisor
        return resultado
    except ZeroDivisionError:
        raise ZeroDivisionError("El divisor no puede ser cero")
```

```python
dividir(1,0)
```
